```python
# Exemplo 4
@widgets.interact_manual(color = ['blue', 'red', 'green'], lw = (1., 10.))
def plot(freq=1., color='blue', lw=2, grid=True):
    t = np.linspace(-1., 1., 1000)
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(t, np.sin(2 * np.pi * freq * t), lw=lw, color=color)
    ax.grid(grid)
```

```python
# Exemplo 5

freq_slider = widgets.FloatSlider(
    value = 2.0, 
    min = 1.0, 
    max = 10.0, 
    step=0.1, 
    description='Frequency:',
    readout_format = '.1f'
)
freq_slider
```

```python
# Exemplo 6
range_slider = widgets.FloatRangeSlider(
    value = [-1., +1.],
    min = -5.0, max = +5.0, step=0.1,
    description = 'xlim:',
    readout_format = '.1f',
)
range_slider
```

```python
# Exemplo 7
grid_button = widgets.ToggleButton(
    value = False,
    description = 'Grid',
    icon = 'check'
)
grid_button
```

```python
# Exemplo 8
color_buttons = widgets.ToggleButtons(
    options = ['blue', 'red', 'green'],
    description = 'Color:',
)
color_buttons
```

```python
# Exmemplo 9
title_textbox = widgets.Text(
    value = 'Hello World',
    description = 'Title:',
) 
title_textbox
```

```python
# Exemplo 10
color_picker = widgets.ColorPicker(
    concise = False,
    description = 'Background color:',
    value = '#efefef',
)
color_picker
```

```python
# Exemplo 11
button = widgets.Button(
    description = 'Plot',
    button_style = 'success',
    icon = 'check'
)
button
```

```python
# Exemplo 12
def plot2(b = None):
    xlim = range_slider.value
    freq = freq_slider.value
    color = color_buttons.value
    grid = grid_button.value
    title = title_textbox.value
    bgcolor = color_picker.value

    t = np.linspace(xlim[0], xlim[1], 1000)
    f, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(t, np.sin(2 * np.pi * freq * t), color=color)
    ax.grid(grid)

@button.on_click
def plot_on_click(b):
    out.clear_output(wait=True)
    with out:
        plot2()
        plt.show()

tab1 = VBox(children=[freq_slider, range_slider])
tab2 = VBox(children=[color_buttons, HBox(children=[title_textbox, color_picker, grid_button])])

out = widgets.Output()
tab = widgets.Tab(children=[tab1, tab2])
tab.set_title(0, 'Plot')
tab.set_title(1, 'Style')
VBox(children=[tab, button, out])
```

```python
# Exemplo 12
from ipywidgets import interact

# Função que calcula o volume da esfera
def volume_esfera(r):
    volume = (4/3) * np.pi * r**3
    print(f"Raio: {r} → Volume da esfera: {volume:.2f} unidades cúbicas")

# Slider interativo para o raio
interact(volume_esfera, r=(0.0, 10.0, 0.1))
```
